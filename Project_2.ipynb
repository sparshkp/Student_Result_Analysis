```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_data.csv")
print(df.head(10))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop the unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
# Gender Ratio Plot
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#from the above chart we can find out that the number of female students are more as compared to the male students

```python
#grouping the  students marks value on the basis of parents education
gp=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.show()
```

From the above chart we can find out that parents education has a greater impact on their children scores

```python
#grouping the marks of students based on the marital status of parents
gp1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp1)
```

```python
sns.heatmap(gp1,annot=True)
plt.show()
```

From the above plot we can conclude that the marital status of parent has no or negligible effect on the student score 

```python
#grouping the marks of students based on the whether they play sports or not 
gp2=df.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp2)
```

```python
sns.heatmap(gp2,annot=True)
plt.show()
```

From the above plot we can find out that there is very less difference between the scores of students who play sports and who don't 

```python
#grouping the marks of students based on the transport means
gp3=df.groupby("WklyStudyHours").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gp3)
```

```python
sns.heatmap(gp3,annot=True)
plt.show()
```

From the plot we can find out that the  number of study hours has very less affect on the student scores

```python
#Distribution of Ethnic groups with pie chart
groupA=df.loc[(df["EthnicGroup"]== "group A")].count()
groupB=df.loc[(df["EthnicGroup"]== "group B")].count()
groupC=df.loc[(df["EthnicGroup"]== "group C")].count()
groupD=df.loc[(df["EthnicGroup"]== "group D")].count()
groupE=df.loc[(df["EthnicGroup"]== "group E")].count()
l=['group A','group B','group C','group D','group E']
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct= "%1.2f%%")
plt.show()
       
```
